```python
# All - QQ plots (Step 4 - "passo 3") (part 1 - preparing)

# max_hours = [14,1,23,23]
# lists = [upload_tv_by_hour, download_tv_by_hour, upload_chr_by_hour, download_chr_by_hour]

# dataset 1 > dataset 3
# dataset 2 < dataset 4

# sorting and creating quantiles ---------------------------------------------------

data_quantiles = [[], [], [], []]

sorted_datasets = [[], [], [], []]


for i in range(4): 

    hour = max_hours[i]

    sorted_datasets[i] = sorted(lists[i][hour]) # sorting the datasets (create new sorted lists, its better)
    
    dataset_size = len(sorted_datasets[i])

    data_quantiles[i] = [i/(dataset_size-1) for i in range(dataset_size)] # creating list of quantiles for each dataset

print(len(sorted_datasets[0]))
print(len(sorted_datasets[1]))
print(len(sorted_datasets[2]))
print(len(sorted_datasets[3]))


# creating interpolation for the qq plot ---------------------------------------------------

def create_interpolation(data_number_x: int, data_number_y: int): # data_number_x is the number of the smaller dataset and data_number_y the bigger
    data_number_x -= 1
    data_number_y -= 1
    
    dataset_x = sorted_datasets[data_number_x]
    dataset_y = sorted_datasets[data_number_y]
    
    size_x = len(dataset_x)
    size_y = len(dataset_y)
    
    quantiles_x = data_quantiles[data_number_x]
    quantiles_y = data_quantiles[data_number_y]
    

    y_list = []

    print(quantiles_x[-1])
    
    for x in quantiles_x:
        if x == 0: # first element
            y_list.append(dataset_y[0])
            continue
        if x == 1: # last element
            y_list.append(dataset_y[-1])
            continue

        for i in range(1,size_y): # finding x1 and x2 (closest quantiles from dataset_y to the x, such that x1 <= x < x2)
            x1 = quantiles_y[i-1]
            x2 = quantiles_y[i]
            
            if (x1<=x<=x2): # determining y1 and y2 (values from dataset_y correspondent to x1 and x2 quantiles)
                y1 = dataset_y[i-1]
                y2 = dataset_y[i]
                break

        y = y1 + (x-x1)*(y2-y1)/(x2-x1) # determining y
        
        y_list.append(y)


    return y_list

sorted_datasets[0] = create_interpolation(3,1) # changing dataset 1, cause dataset 1 > dataset 3
sorted_datasets[3] = create_interpolation(2,4) # changing dataset 4, cause dataset 4 > dataset 2

print(len(sorted_datasets[0]))
print(len(sorted_datasets[1]))
print(len(sorted_datasets[2]))
print(len(sorted_datasets[3]))
```
